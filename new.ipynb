```python
# 小提醒 :

'''
1.抓之前 先確定自己要抓得頁面沒有錯
2. idex 記得更改
3. 清空 temp 
4. adjust the range of page
'''


# 打开数据库连接
db =  pymysql.connect(host='localhost',
                       user='root',
                       passwd='root123456',
                       db='program',
                       charset='utf8')

f = open(r'D:\test\title.txt','w+',encoding = 'utf-8')


# 使用cursor()方法获取操作游标 
cursor = db.cursor()

# original  https://www.ytower.com.tw/recipe/recipe-search.asp?MAINFOOD=%B3J
#         https://www.ytower.com.tw/recipe/pager.asp?MAINFOOD=%B3J&IsMobile=0&page=3

# 中式 0    https://www.ytower.com.tw/recipe/pager.asp?COUNTRY=%A4%A4%A6%A1%AE%C6%B2z&IsMobile=0&page=3
# 韓式 4   https://www.ytower.com.tw/recipe/recipe-search.asp?COUNTRY=%C1%FA%A6%A1%AE%C6%B2z
# 西式 2   https://www.ytower.com.tw/recipe/recipe-search.asp?COUNTRY=%A6%E8%A6%A1%AE%C6%B2z
# 南洋 = 東南亞 9 https://www.ytower.com.tw/recipe/recipe-search.asp?COUNTRY=%ABn%ACv%AE%C6%B2z
# 日式 3 https://www.ytower.com.tw/recipe/recipe-search.asp?COUNTRY=%A4%E9%A6%A1%AE%C6%B2z
# 其他 10 https://www.ytower.com.tw/recipe/recipe-search.asp?COUNTRY=%A8%E4%A5L%AE%C6%B2z
# 冰品 6 https://www.ytower.com.tw/recipe/recipe-search.asp?KIND=%A6B%AB~
# 餅乾麵包 5 https://www.ytower.com.tw/recipe/recipe-search.asp?RETYPE=%BB%E6%B0%AE and
# https://www.ytower.com.tw/recipe/recipe-search.asp?RETYPE=%C4%D1%A5]

baseurl= "https://www.ytower.com.tw/recipe/pager.asp?RETYPE=%C4%D1%A5]&IsMobile=0&page=" 
idex = 52698

# 12 recipes per page 

for i in range(0,15):
    
    sorturl = baseurl + str(i)
    res = requests.get(sorturl)
    res.encoding = 'big5-hkscs'
    print(sorturl)
    soup = BeautifulSoup(res.text,'html.parser')
    #print(soup.prettify())
    
    suggest = soup.find('div',{'id':'content'})
    recipe = suggest.findAll('li')
    
    for eachurl in recipe:
        nonetype = 0;
        tempUrl = eachurl.find('a')['href']
        recipeUrl = "https://www.ytower.com.tw" + tempUrl
        print(recipeUrl) 
        # Start to crawl each recipe
        reciperes = requests.get(recipeUrl)
        reciperes.encoding = 'big5-hkscs'
        recipeSoup = BeautifulSoup(reciperes.text,'html.parser')
        
        
        nonetype = insertRecipe(recipeSoup,idex,recipeUrl)
        if(nonetype == 1 ):
            f.write('%s' % (recipeUrl))
            f.flush()
        else:
            season(recipeSoup,idex)
            step(recipeSoup,idex)
            idex = idex + 1 
        
f.close()
```

```python
def insertRecipe(soup,recode,eachpage):
    
    title = soup.find('div',{'id':'recipe_name'})
    
    if(title == None):
        return 1 
    
    recipe_name = title.find('a').text  # title is nonetype
    print(recipe_name)
    coverdiv = soup.find('div',id="recipe_mainpic")
    cover = coverdiv.find('img')['src']
    #print(cover)
    
    uploadTime = soup.find('time')['datetime']
    #print(uploadTime)
    
    sqlcheck = 'SELECT * FROM receipe WHERE sources = 2 and url = %s'
    cursor.execute(sqlcheck,(eachpage))
    resultUrl = cursor.fetchone()
    if(resultUrl != None):
        recid = resultUrl[0]
        try:
            sql = 'INSERT INTO rec_type(id_rec,id_type) VALUES(%s,%s)'
            cursor.execute(sql,(int(recid),int(5)))
            
            #sqlrecover = 'DELETE FROM rec_type WHERE id_rec = %s and id_type = 4'
            #cursor.execute(sqlrecover,(int(recid)))
            db.commit()
            print("食譜類別更新!")
        except:
            pass
        return 1

    else:
        sql = 'INSERT INTO receipe(id,name,sources,pic,times,url) VALUES(%s,%s,%s,%s,%s,%s)'
        cursor.execute(sql,(int(recode),recipe_name,int(2),cover,uploadTime,eachpage))
        sql = 'INSERT INTO rec_type(id_rec,id_type) VALUES(%s,%s)'
        cursor.execute(sql,(int(idex),int(5)))
        db.commit()
        print("Success insert to 食譜")
        return 0
    
```

```python
def season(soup,recode):
    box = soup.find('div',{'id':'recipe_item'})
    ingredient = box.findAll('ul',{'class':'ingredient'})
    for part in ingredient:
        each = part.findAll('li')
        for detail in each:
            name = ""
            amount = ""
            if(detail.find('span',{'class':'ingredient_name'}) != None):
                name = detail.find('span',{'class':'ingredient_name'}).find('a').text
            if(detail.find('span',{'class':'ingredient_amount'}) != None):
                amount = detail.find('span',{'class':'ingredient_amount'}).text
                
                if(name != "" and amount != ""):
                    print(name + " " + amount)
                    sqlAddFood = 'INSERT INTO temp(id_rec,name,amount) VALUES(%s,%s,%s)'
                    cursor.execute(sqlAddFood,(int(recode),name,amount))
                    db.commit()
```

```python
def step(soup,recode):
    step = soup.findAll('li','step')
    numberStep = 1
    for eachStep in step:
        content = eachStep.text
        print(eachStep.text)
        
        sql2 = 'INSERT INTO steps(id_receipe,step,content) VALUES(%s,%s,%s)'
        cursor.execute(sql2,(int(recode),numberStep,content))
        db.commit()
        numberStep = numberStep + 1
        
    
        
```
