```python
def checkSeasoning(soup,idex):
    dl = soup.findAll('dl','recipe')
    for dd in dl:
        each = dd.find('dd')
        if(each):
            element = re.split("\n", each.text)
            for ingredient in element:
                print(ingredient)
                sqlAddFood = 'INSERT INTO temp(id_rec,name) VALUES(%s,%s)'
                cursor.execute(sqlAddFood,(int(idex),ingredient))
                db.commit()
```

```python
# Insert the each step of the recipe to database
def insertStep(soup,idex):
    stepbox = soup.findAll('div','stepbox')
    for step in stepbox:
        numberStep = step.find('b').text  # number of step
        print(numberStep)   

        if(step.find('img')):    # for the picture 
            stepPic = step.find('img')['src']
            print(stepPic)
        else:
            stepPic = ""      
        
        content = step.find('div',{'class':'word'}).text   # content of step
        print(content)
        
        tips = ""
        if(step.find('p',{'class':'tips'})!= None):
            print("小技巧" + step.find('p',{'class':'tips'}).text)
            tips = step.find('p',{'class':'tips'}).text
            

        
        try:
            sql2 = 'INSERT INTO steps(id_receipe,step,content,pic) VALUES(%s,%s,%s,%s)'
            cursor.execute(sql2,(int(idex),numberStep,content,stepPic))
            db.commit()
            sqltip = 'INSERT INTO tip(id_rec,count,content) VALUES (%s,%s,%s)'
            cursor.execute(sqltip,(int(idex),numberStep,tips))
            db.commit()
        except:
            pass
```

```python
def hashtag(soup,idex,auto):
    hashtag = soup.find('div','kwtab boxTitle boxText')
    for atag in hashtag.find_all('a'):
        # print(atag.text)
        tagname = atag.text
        sqlcheck = 'SELECT * FROM cate_receipe WHERE name = %s'
        cursor.execute(sqlcheck,(tagname))
        resultTag = cursor.fetchone()

        sqltag = 'INSERT INTO rec_cate(id_rec,id_cate) VALUES(%s,%s)'


        if(resultTag != None):
            tagId = resultTag[0]
            sqlduplicate = 'SELECT * FROM rec_cate WHERE id_rec = %s and id_cate = %s '
            cursor.execute(sqlduplicate,(int(idex),int(tagId)))
            duplicateTag = cursor.fetchone()
            if(duplicateTag == None):
                cursor.execute(sqltag,(int(idex),tagId))
                db.commit()

        elif(resultTag == None):
            sqlInsertTag = 'INSERT INTO cate_receipe(id,name) VALUES(%s,%s)'
            cursor.execute(sqlInsertTag,(int(auto),tagname))
            db.commit()
            sqltag = 'INSERT INTO rec_cate(id_rec,id_cate) VALUES(%s,%s)'
            cursor.execute(sqltag,(int(idex),int(auto)))
            db.commit()
            auto = auto + 1
            
    return auto
```

```python
# Find the Recipe Title and Cover
def insertRecipe(eachpage,soup,idex,coverUrl):
    
    div = soup.find('div','print_re')
    
    regex = re.compile(r".youtube.")
    reface = re.compile(r"https://www.facebook.com/plugins/video.php?.*width.")
    
    youtube = soup.findAll('iframe')
    videoUrl = ""
    for check in youtube:
        if(regex.findall(check['src'])):
            #print(check['src'])
            videoUrl = check['src']
        elif(reface.findall(check['src'])):
            videoUrl = check['src']

    
    
    title = soup.find('h1').text
    
    content = soup.find('div',{'class':'context cookbook'})
    if(content == None):
        return 1
    ptag = content.findAll('p',{'class':None}, recursive=False)
    introduction = ""
    for te in ptag:
        introduction += te.text
    print(introduction)
        
    textbook = soup.findAll("p")   # introduction 這邊忽略了一些各個作者不同的事前準備之類的
    introduce = textbook[0].text
    print(textbook[0].text)
    
    time = soup.find('span','author').find('b').text
    print(time)
    
   
    sqlcheck = 'SELECT * FROM receipe WHERE sources = 1 and url = %s'
    cursor.execute(sqlcheck,(eachpage))
    resultUrl = cursor.fetchone()
    if(resultUrl != None):
        recid = resultUrl[0]
        try:
            sql = 'INSERT INTO rec_type(id_rec,id_type) VALUES(%s,%s)'
            cursor.execute(sql,(int(recid),int(15)))
            db.commit()
            print("食譜類別更新!")
        except:
            pass
        return 1
    else:
        sql = 'INSERT INTO receipe(id,name,sources,pic,introduce,times,video,url) VALUES(%s,%s,%s,%s,%s,%s,%s,%s)'
        cursor.execute(sql,(int(idex),title,int(1),coverUrl,introduction,time,videoUrl,eachpage))
        sql = 'INSERT INTO rec_type(id_rec,id_type) VALUES(%s,%s)'
        cursor.execute(sql,(int(idex),int(15)))
        
        db.commit()
        print("Success insert to 食譜")
        return 0
    
```

```python
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 
```
